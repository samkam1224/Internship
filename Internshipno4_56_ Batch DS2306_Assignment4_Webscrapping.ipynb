```python
from bs4 import BeautifulSoup
import requests
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
titles = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
```

```python
print('List all the header tags :', *titles, sep='\n\n')
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':titles})
```

```python
from bs4 import BeautifulSoup
import requests
page1=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
```

```python
soup1 = BeautifulSoup(page1.content, 'html.parser')
```

```python
soup1
```

```python
name=soup1.find('h3.div',class_='presidentListing')
name
```

```python
Term_of_Office=soup1.find('#text',class_='terms')
Term_of_Office
```

```python
names=[]
for i in soup1.find_all('div',class_='presidentListing'):
    names.append(i.get('h3.div'))

names
```

```python
Term_of_Offices=[]
for i in soup1.find_all('span',class_='terms'):
    Term_of_Offices.append(i.get('span'))

Term_of_Offices
```
